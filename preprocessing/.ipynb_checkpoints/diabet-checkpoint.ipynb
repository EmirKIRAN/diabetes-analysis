# Eksik (NaN) değerlerin incelenmesi
Veri setine ait özniteliklerde bazı değerlerin eksik olup olmadığına bakılmalıdır. Burada eksik değerler 0 olarak belirtilmiştir. Bunların düzeltilmesi gerekir

```python
diabetes_data_copy = diabetes_data.copy(deep=True)
diabetes_data_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = diabetes_data_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
print(diabetes_data_copy.isnull().sum())
```

# Çıkarımlar

 * BloodPressure, Glucose, Insulin, BMI, SkinThickness özniteliklerinde aykırı değerler(outlier) görülmektedir.
 * Outlier değerler genel olarak 0 dan oluşmaktadır.
 * Outcome özniteliğinin çoğunluğunu 0 değeri oluşturmaktadır.

# Eksik verilerin doldurulması

```python
def fillNaN(dataset,column,value):
    dataset[column].fillna(value,inplace=True)
```

```python
fillNaN(diabetes_data_copy,'Glucose',diabetes_data_copy['Glucose'].mean())
fillNaN(diabetes_data_copy,'BloodPressure',diabetes_data_copy['BloodPressure'].mean())
fillNaN(diabetes_data_copy,'SkinThickness',diabetes_data_copy['SkinThickness'].median())
fillNaN(diabetes_data_copy,'Insulin',diabetes_data_copy['Insulin'].median())
fillNaN(diabetes_data_copy,'BMI',diabetes_data_copy['BMI'].mean())
```

# Normalizsyon İşlemi

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X =  pd.DataFrame(sc_X.fit_transform(diabetes_data_copy.drop(["Outcome"],axis = 1),),
        columns=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age'])
```

```python
X.head()
```

```python
y = diabetes_data_copy.Outcome
```

# Eğitim ve Test verilerinin ayrılması

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42,stratify=y)
```

# KNN Algoritması ile eğitim

Aşağıda en yüksek skora sahip model kullanılmak üzere seçilmiştir.

```python
from sklearn.neighbors import KNeighborsClassifier

test_scores = []
train_scores = []

for i in range(1,15):
    knn = KNeighborsClassifier(i)
    knn.fit(X_train,y_train)
    
    train_scores.append(knn.score(X_train,y_train))
    test_scores.append(knn.score(X_test,y_test))
```

```python
max_train_score = max(train_scores)
train_score_ind = [i for i,v in enumerate(train_scores) if v == max_train_score]
print('Max train score {} % and k = {}'.format(max_train_score*100,list(map(lambda x: x+1, train_score_ind))))
```

```python
max_test_score = max(test_scores)
test_scores_ind = [i for i,v in enumerate(test_scores) if v == max_test_score]
print('Max train score {} % and k = {}'.format(max_test_score*100,list(map(lambda x: x+1, test_scores_ind))))
```

```python
plt.figure(figsize=(12,5))
p = sns.lineplot(range(1,15),train_scores,marker="*",label="Train Score")
p = sns.lineplot(range(1,15),test_scores,marker="o",label="Test Score")
```

### Yukarıdaki grafiğe bakarak en iyi parametre değerinin 5 olduğunu görebiliriz

# Modelin parametrelerinin güncellenmesi ve eğitilmesi

```python
knn = KNeighborsClassifier(5)
knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```

# Confusion Matrix ile tahmin ölçümü

```python
y_pred = knn.predict(X_test)
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
p = sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

## Sonuç

 * Model %74'lük bir tahmin başarımı elde etmiştir.
 * Model 143 adet doğru tahmin, 49 adet hatalı tahmin yapmıştır
